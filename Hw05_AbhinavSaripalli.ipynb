```python
#PROBLEM 1
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
labels=[]
data_path='C://Users//Abhinav//Desktop//student_class.csv'
data=pd.read_csv(data_path)
data1=data[['Midterm', 'Final', 'Final Average', 'Majors']]
lst=data1['Majors'].unique()
data3={}
for major in lst:
    data2=data1.loc[data1['Majors']==major]
    if major not in data3:
        labels.append(major)
        data3[major]=data2.median()
data4=pd.DataFrame.from_dict(data3)
data5=data4.T
x=np.arange(len(labels))
width=0.25
fig, ax = plt.subplots(figsize=(10, 7))
rects1 = ax.bar(x, data5['Midterm'], width, label='Midterm', color='r')
rects2 = ax.bar(x + width, data5['Final'], width, label='Final', color='g')
rects3 = ax.bar(x + 2*width, data5['Final Average'], width, label='Final Average', color='b')
ax.set_xlabel('Major')
ax.set_ylabel('Median Score')
ax.set_title('Median Midterm, Final and Final Average Scores')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc='upper right')
plt.show()
```

```python
#PROBLEM 3
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
data_path='C://Users//Abhinav//Desktop//student_class.csv'
data=pd.read_csv(data_path)
ax= data.plot(kind='scatter', x='Midterm', y='Final Average', label='Midterm', color='red')
data.plot(kind='scatter', x='Final', y='Final Average', label='Final', color='blue', ax=ax)
ax.legend()
plt.title('Exams vs Final Average')
plt.xlabel('Exam Score')
plt.ylabel('Final Average')
plt.gcf().set_size_inches(12, 5)
plt.show()
```
